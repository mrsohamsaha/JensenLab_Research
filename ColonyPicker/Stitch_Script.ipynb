```python
import cv2
import imutils
from matplotlib import pyplot as plt
```

```python
def showImages(images):
    count = 0
    for img in images:
        windowName = "img" + str(count)
        cv2.namedWindow(windowName, cv2.WINDOW_NORMAL)
        cv2.imshow(windowName,img)
        cv2.waitKey(0)
        count += 1
    cv2.destroyAllWindows()
```

```python
topRight = cv2.imread('topRight.jpg')
topRight = imutils.rotate(topRight, -3)
topRight = topRight[370:1940, 350:1910]
plt.imshow(topRight, 'gray')
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
print("x: " + str(topRight.shape[1]) + " y: " + str(topRight.shape[0]))
# showImages([topRight])
```

```python
cv2.imwrite('TR.jpg',topRight)
```
